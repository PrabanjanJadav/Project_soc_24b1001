Problem solving using basic python,numpy,matplotlib and pandas


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('tmdb_5000_movies.csv')

def row(part):
    part = part.strip('[]')
    if part == '':
        return []
    parts = part.split("}, {")
    genres=[]
    for thing in parts:
        start = thing.find('"name": "') + len('"name": "')
        end = thing.find('"', start)
        genre_name= thing[start:end]
        genres.append(genre_name)
    
    return genres
all_genres = []
for thing in df['genres']:
    all_genres.append(row(thing))

print (all_genres)
each_genre = np.concatenate(all_genres)

counts = {}

for item in each_genre:
    if item in counts:
        counts[item] += 1
    else:
        counts[item] = 1
print(each_genre)
print(counts)
genres= list(counts.keys())
repetitions = list(counts.values())

plt.bar(genres, repetitions)
plt.xticks(rotation=45, ha='right')
plt.title('Number of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')

plt.tight_layout()
plt.savefig('genres_plot.png')  
plt.show() 
```
